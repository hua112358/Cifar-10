```python
"""
    Step 1: Build Model

"""

class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.conv1 = torch.nn.Conv2d(3, 32, 5)
        self.pool1 = torch.nn.MaxPool2d(2, 2)
        self.conv2 = torch.nn.Conv2d(32, 64, 5)
        self.pool2 = torch.nn.MaxPool2d(2, 2)
        self.fc1 = torch.nn.Linear(64 * 5 * 5, 1024)
        self.fc2 = torch.nn.Linear(1024, 256)
        self.fc3 = torch.nn.Linear(256, 10)
#         self.fc1 = torch.nn.Linear(64 * 53 * 53, 10)
    def forward(self, input):
        layer1 = torch.functional.F.relu(self.conv1(input))
        layer2 = self.pool1(layer1)
        layer3 = torch.functional.F.relu(self.conv2(layer2))
        layer4 = self.pool2(layer3)
        layer5 = layer4.view(-1, 64 * 5 * 5)
        layer6 = torch.functional.F.relu(self.fc1(layer5))
        layer7 = torch.functional.F.relu(self.fc2(layer6))
        output = self.fc3(layer7)
#         output = self.fc1(layer5)
        return output
    
# transfer learning with resnet18
# 1 transfer learning with better weight initialization
resnet18 = torchvision.models.resnet18(pretrained=True)

# resnet18.conv1 = torch.nn.Conv2d(3, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False)
resnet18.maxpool = torch.nn.MaxPool2d(2, 2, 1, 1)
resnet18.avgpool = torch.nn.AvgPool2d(2, 1, 0, False, True)
resnet18.fc = torch.nn.Linear(resnet18.fc.in_features, 10)

# # only use the architecture without pretrained weight
# l = list(resnet18.children())
# resnet18 = torch.nn.Sequential(*(l[:3] + l[4:])) # drop maxpool

# 2 transfer learning as feature extractor
resnet18 = torchvision.models.resnet18(pretrained=True)

# resnet18.conv1 = torch.nn.Conv2d(3, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False)
resnet18.maxpool = torch.nn.MaxPool2d(2, 2, 1, 1)
resnet18.avgpool = torch.nn.AvgPool2d(2, 1, 0, False, True)
for param in resnet18.parameters():
    param.requires_grad = False
# list(resnet18.parameters())[0].requires_grad = True
# list(resnet18.parameters())[1].requires_grad = True
resnet18.fc = torch.nn.Linear(resnet18.fc.in_features, 10)

# # only use the architecture without pretrained weight
# l = list(resnet18.children())
# resnet18 = torch.nn.Sequential(*(l[:3] + l[4:])) # drop maxpool

# model = resnet18
model = Model()

print("total parameters: %d" % sum([param.numel() for param in model.parameters()]))
print("parameters: ", [param.numel() for param in model.parameters()])
print("model: ", model)
```
